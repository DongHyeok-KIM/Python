```python
import folium
from pandas import DataFrame #데이터 분석 패키지
from pandas import ExcelFile #엑셀 파일 가져오기
```

##  #03. 서울의 초등학교 위치
### 1) 데이터 수집

```python
# 엑셀 파일을 원경으로 내려받아 데이터 프레임으로 변환(다소 시간이 소요된다.)
xlsx = ExcelFile("http://itpaper.co.kr/demo/py/school2019.xlsx")
df = xlsx.parse(xlsx.sheet_names[0])
df
```

## #02. 데이터 전처리
### 1) 사용할 필드만 추출

```python
df2 = df.filter(['학교명', '학교급구분', '소재지도로명주소', '위도', '경도'])
df2
```

## 2) 서울시의 초등학교만 추출
1. '학교급구분'필드 값이 **초등학교**
1. '소재지도로명주소'에 **서울** 이라는 단어가 포함된 경우
    '''
    python
    #### LIKE연산
    컬럼이름.str.contains('검색어')
    '''

```python
df3 = df2.query("학교급구분 == '초등학교' and 소재지도로명주소.str.contains('서울')")
df3
```

## #03. 데이터 시각화
### #01) 지도객체 생성

```python
#zoom_start: 배율 1~22
map_osm = folium.Map(location=[37.566651, 126.978428], zoom_start=12)
map_osm
```

```python
## 02)데이터 프레임의 행 수만큼 반복하면서 마커 생성
```

```python
for i in df3.index:
    #행 우선 접근 방식으로 값 추출하기
    name = df3.loc[i, '학교명']
    lat = df3.loc[i, '위도']
    lng = df3.loc[i,'경도']
    
    # 추출한 정보를 지도에 표시
    marker = folium.Marker([lat,lng], popup=name)
    marker.add_to(map_osm)
map_osm
```
